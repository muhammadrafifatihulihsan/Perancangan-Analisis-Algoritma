```python
import pandas as pd
```

```python
# MEMBACA DATASET
df = pd.read_csv("../dataset/ecommerce_data.csv")

print("DATASET")
print(df.head())
```

```python
# FUNGSI JACCARD SIMILARITY
# Rumus: J(A,B) = intersection(A,B) / union(A,B)
def jaccard_similarity_manual(set_a, set_b):
    
    # INTERSECTION
    # User yang membeli kedua produk
    intersection = set_a.intersection(set_b)

    # UNION
    # User yang membeli salah satu
    # atau kedua produk
    union = set_a.union(set_b)

    similarity = len(intersection) / len(union)
    print(f'similarity: {similarity}')
    return similarity
```

```python
# MEMILIH PRODUK
product_a = "P101"
product_b = "P108"
```

```python
# MENGAMBIL USER DARI MASING-MASING PRODUK
users_a = set(
    df[df['Product_ID'] == product_a]['User_ID']
)

users_b = set(
    df[df['Product_ID'] == product_b]['User_ID']
)
```

```python
# DETAIL PRODUK
detail_a = df[df['Product_ID'] == product_a].iloc[0]
detail_b = df[df['Product_ID'] == product_b].iloc[0]

print("\nDETAIL PRODUK")

print(f"""
Product ID   : {detail_a['Product_ID']}
Nama Produk  : {detail_a['Product_Name']}
Jumlah User  : {len(users_a)}
""")

print(f"""
Product ID   : {detail_b['Product_ID']}
Nama Produk  : {detail_b['Product_Name']}
Jumlah User  : {len(users_b)}
""")
```

```python
# INTERSECTION DAN UNION
intersection = users_a.intersection(users_b)
union = users_a.union(users_b)

print("\nUSER INTERSECTION")
print(intersection)

print("\nUSER UNION")
print(union)

print("\nJumlah Intersection =", len(intersection))
print("Jumlah Union =", len(union))
```

```python
# HITUNG JACCARD
similarity = jaccard_similarity_manual(users_a, users_b)
```

```python
# MENAMPILKAN HASIL
print("\nHASIL JACCARD SIMILARITY")

print(
    f"{product_a} vs {product_b} =",
    round(similarity, 4)
)
```

```python
# REKOMENDASI PRODUK
print("\nPRODUK REKOMENDASI")

print(f"""
Jika user membeli:
{detail_a['Product_Name']}

Maka sistem merekomendasikan:
{detail_b['Product_Name']}

Karena memiliki nilai Jaccard Similarity
sebesar {round(similarity,4)}
""")
```

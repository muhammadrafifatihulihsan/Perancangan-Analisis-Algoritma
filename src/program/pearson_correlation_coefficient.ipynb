```python
import pandas as pd
import numpy as np
```

```python
# MEMBACA DATASET
df = pd.read_csv("../dataset/ecommerce_data.csv")

print("DATASET")
print(df.head())
```

```python
# MEMBENTUK USER-ITEM MATRIX
matrix = df.pivot_table(
    index='User_ID',
    columns='Product_ID',
    values='Rating'
)

print("\nUSER ITEM MATRIX")
print(matrix)
```

```python
# FUNGSI PEARSON CORRELATION
# Parameter:
# vec_a = vektor user pertama
# vec_b = vektor user kedua
#
# Rumus: PCC = sum((a-a_mean)*(b-b_mean)) / sqrt(sum((a-a_mean)^2)) * sqrt(sum((b-b_mean)^2))

def pearson_correlation_manual(vec_a, vec_b):
    # RATA-RATA
    mean_a = np.mean(vec_a)
    mean_b = np.mean(vec_b)

    # DEVIASI DARI RATA-RATA
    dev_a = vec_a - mean_a
    dev_b = vec_b - mean_b

    # PEMBILANG
    numerator = np.sum(dev_a * dev_b)
    
    # PENYEBUT
    denominator = (
        np.sqrt(np.sum(dev_a ** 2))
        *
        np.sqrt(np.sum(dev_b ** 2))
    )

    # HANDLE PEMBAGIAN 0
    if denominator == 0:
        return 0
    
    similarity = numerator / denominator
    print(f"similarity: {similarity}")
    return similarity
```

```python
# MEMILIH USER
user_a = "U001"
user_b = "U002"
```

```python
# MENCARI PRODUK YANG SAMA-SAMA DIRATING
common_items = matrix.loc[[user_a, user_b]].dropna(axis=1)

print("\nPRODUK YANG DIRATING BERSAMA")
print(common_items)
```

```python
# MENGAMBIL VECTOR RATING

vec_a = common_items.loc[user_a].values
vec_b = common_items.loc[user_b].values

print("\nVEKTOR RATING")

print(f"{user_a} =", vec_a)
print(f"{user_b} =", vec_b)
```

```python
# DETAIL INFORMASI PRODUK
print("\nDETAIL PRODUK YANG DIBANDINGKAN")

for product_id in common_items.columns:
    product_name = df[df['Product_ID'] == product_id]['Product_Name'].iloc[0]
    rating_a = common_items.loc[user_a, product_id]
    rating_b = common_items.loc[user_b, product_id]
    
    print(f"""
Product ID    : {product_id}
Nama Produk   : {product_name}
Rating {user_a} : {rating_a}
Rating {user_b} : {rating_b}
""")
```

```python
# PERHITUNGAN MANUAL
mean_a = np.mean(vec_a)
mean_b = np.mean(vec_b)

dev_a = vec_a - mean_a
dev_b = vec_b - mean_b

numerator = np.sum(dev_a * dev_b)

denominator = (
    np.sqrt(np.sum(dev_a ** 2))
    *
    np.sqrt(np.sum(dev_b ** 2))
)

similarity = pearson_correlation_manual(vec_a, vec_b)
```

```python
# MENAMPILKAN HASIL
print("\nHASIL PERHITUNGAN")

print(f"Rata-rata {user_a} =", round(mean_a, 4))
print(f"Rata-rata {user_b} =", round(mean_b, 4))

print(f"\nDeviasi {user_a} =", dev_a)
print(f"Deviasi {user_b} =", dev_b)

print("\nPembilang =", round(numerator, 4))
print("Penyebut =", round(denominator, 4))

print("\nHASIL PEARSON CORRELATION")
print(f"{user_a} vs {user_b} =", round(similarity, 4))
```

```python
# MENCARI PRODUK REKOMENDASI
produk_user_a = set(
    df[df['User_ID'] == user_a]['Product_ID']
)

produk_user_b = set(
    df[df['User_ID'] == user_b]['Product_ID']
)

rekomendasi = produk_user_b - produk_user_a

print("\nPRODUK REKOMENDASI")

for product_id in rekomendasi:

    detail = df[df['Product_ID'] == product_id].iloc[0]

    print(f"""
Product ID   : {detail['Product_ID']}
Nama Produk  : {detail['Product_Name']}
Rating User  : {detail['Rating']}
""")
```
